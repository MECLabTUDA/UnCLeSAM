```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# Define custom colormap colors
custom_colors = ['#9e1e83', '#fae5d6', '#fec10b', '#26a671' ]

# Create custom colormap
custom_cmap = LinearSegmentedColormap.from_list("custom", custom_colors)

# Data for UnCLe SAM and nnUNet
uncle_sam_data = np.array([
    [81.55, 59.94, 79.77, 92.27],
    [84.68, 54.84, 96.02, 82.56],
    [85.99, 88.11, 83.55, 80.16],
    [85.29, 51.53, 85.57, 44.57]
])

nnunet_data = np.array([
    [25.24, 27.57, 59.2, 89.25],
    [81.78, 29.06, 93.0, 52.48],
    [57.11, 83.06, 45.73, 1.3],
    [86.6, 32.6, 86.69, 3.49]
])

methods = ['UnCLe SAM', 'nnUNet']
datasets = ['DecathProst', 'ISBI', 'I2CVB', 'UCL']

# Determine common color scale limits
vmin = min(np.min(uncle_sam_data), np.min(nnunet_data))
vmax = max(np.max(uncle_sam_data), np.max(nnunet_data))

# Plot confusion matrices with custom colormap
plt.figure(figsize=(12, 5))

# UnCLe SAM
plt.subplot(1, 2, 1)
sns.heatmap(uncle_sam_data, annot=True, fmt='.2f', cmap=custom_cmap, xticklabels=datasets[::-1], yticklabels=datasets, vmin=vmin, vmax=vmax)
plt.title('UnCLe SAM - Training Confusion Matrix')

# nnUNet
plt.subplot(1, 2, 2)
sns.heatmap(nnunet_data, annot=True, fmt='.2f', cmap=custom_cmap, xticklabels=datasets[::-1], yticklabels=datasets, vmin=vmin, vmax=vmax)
plt.title('nnUNet - Training Confusion Matrix')

plt.tight_layout()
# plt.show()
plt.savefig(f'/media/aranem_locale/AR_subs_exps/MIDL_2024/MIDL_2024_evaluation/plots/confusion_matrix.pdf', bbox_inches='tight')
```
